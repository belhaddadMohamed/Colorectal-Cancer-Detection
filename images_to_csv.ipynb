```python
import os
import cv2
import numpy as np
import pandas as pd
```

```python

# Define the image size
IMG_SIZE = 224
```

```python

# Define the class labels
class_labels = ['ADI', 'BACK', 'DEB', 'LYM', 'MUC', 'MUS', 'NORM', 'STR', 'TUM']

```

```python
# Load the images and store the pixel values in a numpy array
data = []
labels = []
for class_label in class_labels:
    path = os.path.join('C:/Users/Sc/Desktop/Colorectal Cancer/Dataset/CRC-VAL-HE-7K', class_label)
    for img_file in os.listdir(path):
        img_path = os.path.join(path, img_file)
        img = cv2.imread(img_path)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img.astype('float32') / 255.0  # normalize pixel values between 0 and 1
        data.append(img.flatten())
        labels.append(class_label)

data = np.array(data)
labels = np.array(labels)

```

```python
# Combine the data and labels into a pandas dataframe
df = pd.DataFrame(data)
df['label'] = labels

```
